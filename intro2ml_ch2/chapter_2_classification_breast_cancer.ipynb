```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
import numpy as np
```

```python
cancer = load_breast_cancer()
```

```python
print(f"Cancer keys: {cancer.keys()}")
```

```python
print(f"Cancer data shape:{cancer.data.shape}")
```

```python
print(f"Cancer data: {cancer.data}")
```

```python
print(f"Cancer target shape: {cancer.target.shape}")
```

```python
print(f"Cancer target:\n {cancer.target}")
```

```python
print(f"Cancer target names shape: {cancer.target_names.shape}")
```

```python
print(f"Cancer target names: \n {cancer.target_names}")
```

```python
class_names = cancer.target_names
sample_count_per_class = np.bincount(cancer.target)
class_zip = zip(class_names,sample_count_per_class)

for k,v in class_zip:
    print(k,":",v)
```

```python
print(f"Cancer feature_names:\n {cancer.feature_names}" )
```

```python
X_train,X_test,y_train,y_test = train_test_split(cancer.data,cancer.target,stratify = cancer.target,random_state=33)
```

```python
training_acc = []
test_acc = []
k_range = range(1,11)
for n in k_range:
    clf = KNeighborsClassifier(n_neighbors=n)
    clf.fit(X_train,y_train)
    training_acc.append(clf.score(X_train,y_train))
    test_acc.append(clf.score(X_test,y_test))

plt.plot(k_range,training_acc,label="Training Accuracy")
plt.plot(k_range,test_acc,label="Test Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("n_neighbors")
plt.legend(loc='best')

print("\t Accuracy")
for k,v in zip(training_acc,test_acc):
    print(f"Train: {k:.2f} | Test: {v:.2f}")
```

# Kernel Trick SVM

```python
from sklearn.svm import SVC
rbf = SVC() #rbf kernel svm (radial basis function) 
# SVC default C=1, gamma = 1/n_Features
```

```python
rbf.fit(X_train,y_train)
print("Train Acc:",rbf.score(X_train,y_train))
print("Test Acc:",rbf.score(X_test,y_test))
```

```python
my_gamma= [0.01,1,10000]
my_c = [0.1,1,1000]
result = []

for i in range(3):
    for j in range(3):
        rbf = SVC(C=my_c[i],gamma=my_gamma[j]).fit(X_train,y_train)
        train_acc = rbf.score(X_train,y_train)
        test_acc = rbf.score(X_test,y_test)
        key = f"C={my_c[i]},Gamma={my_gamma[j]},Train Acc:{train_acc:.2f},Test Acc:{test_acc:.2f}"
        result.append(key)

for i in result:
    print(i)
```

```python
plt.plot(X_train.min(axis=0), 'o', label="min")
plt.plot(X_train.max(axis=0), '^', label="max")
plt.legend(loc=4)
plt.xlabel("Feature index")
plt.ylabel("Feature magnitude")
plt.yscale("log")
```

# Preprocessing data for SVM : Rescaling (MinMaxScaler)

```python
# compute the minimum value per feature on the training set
min_on_training = X_train.min(axis=0)
# compute the range of each feature (max - min) on the training set
range_on_training = (X_train - min_on_training).max(axis=0)
# subtract the min, and divide by range
# afterward, min=0 and max=1 for each feature
X_train_scaled = (X_train - min_on_training) / range_on_training
print("Minimum for each feature\n{}".format(X_train_scaled.min(axis=0)))
print("Maximum for each feature\n {}".format(X_train_scaled.max(axis=0)))
```

```python
X_test_scaled = (X_test-min_on_training)/range_on_training
```

```python
svc = SVC().fit(X_train_scaled,y_train)
print("Train Acc:",svc.score(X_train_scaled,y_train))
print("Test Acc:",svc.score(X_test_scaled,y_test))
```

```python
svcComplex = SVC(C=2,gamma=0.1).fit(X_train_scaled,y_train)
print("Train Acc:",svcComplex.score(X_train_scaled,y_train))
print("Test Acc:",svcComplex.score(X_test_scaled,y_test))
```

```python
# RBF ile SVC kurduğumuzda %91'e %92 sonuçlarını almışken, scaling ile beraber bu sonuç %98 e %97 oldu. C ve gamma değerleriyle oynayarak %97 'e %98 olarak başarı oranı yakalayabildik.'
```

# Derin Öğrenme: MultiLayerPerceptron

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(random_state=42)
```

```python
mlp.fit(X_train,y_train)
```

```python
print("Train Acc:",mlp.score(X_train,y_train))
print("Test Acc:",mlp.score(X_test,y_test))
```

```python
mean_on_train = X_train.mean(axis=0)
std_on_train = X_train.std(axis=0)

X_train_scaled = (X_train - mean_on_train) / std_on_train
X_test_scaled = (X_test - mean_on_train) / std_on_train
```

```python
mlp_scaled = MLPClassifier(random_state=42).fit(X_train_scaled,y_train)
print("Train Acc:",mlp_scaled.score(X_train_scaled,y_train))
print("Test Acc:",mlp_scaled.score(X_test_scaled,y_test))
```

```python
# Uyarı diyorki iterasyon sayısına maximuma ulaştı. Iterasyon sayısını arttırmalıyız.
```

```python
mlp_scaled = MLPClassifier(random_state=42,max_iter=300).fit(X_train_scaled,y_train)
print("Train Acc:",mlp_scaled.score(X_train_scaled,y_train))
print("Test Acc:",mlp_scaled.score(X_test_scaled,y_test))
```

```python
mlp_scaled = MLPClassifier(random_state=42,max_iter=300,alpha=1).fit(X_train_scaled,y_train) # lr default olarak 1e-4 tü onu 1'e çektim..
print("Train Acc:",mlp_scaled.score(X_train_scaled,y_train))
print("Test Acc:",mlp_scaled.score(X_test_scaled,y_test))
```
